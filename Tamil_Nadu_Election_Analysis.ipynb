```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Tn_Election_2026.xlsx")

df.head()
```

```python
print(df.columns)
```

```python
tvk_df = df[df["party"] == "TVK"]

tvk_df.head()
```

```python
regional_votes = tvk_df.groupby("region")["votes"].sum()
print(regional_votes)
```

```python
regional_votes = tvk_df.groupby("region")["votes"].mean()
print(regional_votes)
```

```python
regional_votes.plot(kind='bar')

plt.title("Average TVK Votes Across Regions")
plt.xlabel("Region")
plt.ylabel("Average Votes")

plt.show()
```
